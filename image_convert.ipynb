# Convert pills image background to transparent
This notebook removes the white background from the target image and saves a transparent PNG that matches the login panel background.

```python
# Import Required Libraries
import os
from PIL import Image
import numpy as np
print('libraries imported')
```

```python
# Load the Images
base_dir = os.path.abspath(os.path.dirname(__file__))
target_path = os.path.join(base_dir, 'public', 'images', 'pills.jpg')
output_path = os.path.join(base_dir, 'public', 'images', 'pills-transparent.png')
print('target image:', target_path)
img = Image.open(target_path).convert('RGBA')
img_np = np.array(img)
print('loaded image size:', img_np.shape)
```

```python
# Extract Background Color from Reference Image
background_color = (20, 85, 210, 255)  # representative blue
print('reference background color:', background_color)
```

```python
# Remove White Background from Target Image
threshold = 240
transparent_img = img.copy()
pixels = transparent_img.load()
for y in range(transparent_img.height):
    for x in range(transparent_img.width):
        r, g, b, a = pixels[x, y]
        if r > threshold and g > threshold and b > threshold:
            pixels[x, y] = (255, 255, 255, 0)
print('white background made transparent')
```

```python
# Apply New Background Color
transparent_img.save(output_path)
preview = Image.new('RGBA', transparent_img.size, background_color)
preview.paste(transparent_img, (0, 0), transparent_img)
preview_path = os.path.join(base_dir, 'public', 'images', 'pills-preview.png')
preview.save(preview_path)
print('saved transparent image:', output_path)
print('saved preview image:', preview_path)
transparent_img
```
